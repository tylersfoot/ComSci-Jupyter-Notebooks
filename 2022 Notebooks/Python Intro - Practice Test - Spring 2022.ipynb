3. Using a for loop, find the sum of the multiples of both 3 and 4, starting from 1 and checking all integers up to 500. No math shortcuts here either.

```python
sum = 0
for i in range(1, 500):
    if i%4==0 and i%3==0:
        sum+=i
print(sum)
```

5. Write a while loop that prints out the even integers up to 100, but skips multiples of 10 by using a break or continue statement.

```python
for i in range(2, 100, 2):
    if i%10==0:
        continue
    print(i, end=' ')
```
