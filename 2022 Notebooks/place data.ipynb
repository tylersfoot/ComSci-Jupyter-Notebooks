```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
df = pd.read_csv('./data/placedata/individual/place7.csv', sep=',')
# dfFull = pd.read_csv('./data/placedata/placefull.csv', sep=',')
```

```python
df.head(5)
```

```python
# del df['user_id']
# del df['timestamp']
# del df['pixel_color']
coordlist = df.coordinate.values.tolist()
coordlist
x = []
y = []
for i in range(len(coordlist)):
    a, b, = map(float, coordlist[i].strip('()').split(','))
    x.append(a)
    y.append(b)
dfCoords=pd.DataFrame({'x': x, 'y': y})
```

```python
dfCoords = dfCoords[::20]
```

```python
# set seaborn style
sns.set_style("white")

# Basic 2D density plot
sns.kdeplot(x=dfCoords.x, y=dfCoords.y)
plt.show()

# Add thresh parameter
sns.kdeplot(x=dfCoords.x, y=dfCoords.y, cmap="Blues", shade=True, thresh=0)
plt.show()
```

```python
sns.kdeplot(x=dfCoords.x, y=dfCoords.y, cmap="Reds", shade=True, bw_adjust=.01, thresh=0)
plt.show()
```
