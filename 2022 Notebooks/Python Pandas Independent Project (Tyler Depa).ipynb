# Osu! Top 250 Players Dataset

### This dataset contains the top 250 ranked players in the rythm video game 'osu!', including their username, levels they played (play_count), total performance points (method of determining how good a player is), their average accuracy across levels (%), and the amount of the different accuracies they scored (SS [100%], S [95%], A [90%])

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
plt.style.use('ggplot')

df = pd.read_csv('./data/osu.csv', sep=',')
df['rank'] = df['rank'].str.replace('#','').astype(int) # Converts rank (#14) to int (14)
df['accuracy'] = df['accuracy'].str.replace('%','').astype(float) # Converts accuracy (96.1%) to float (96.1)
df['A'] = df['A'].apply(lambda x: (x*1000 if int(x) != x else x)) # Converts from european notation (1.320) to normal (1320)
df['S'] = df['S'].apply(lambda x: (x*1000 if int(x) != x else x))
df['SS'] = df['SS'].apply(lambda x: (x*1000 if int(x) != x else x))
df['performance'] = df['performance'] * 1000 # These two were counted in thousands, converted to raw numbers
df['play_count'] = df['play_count'] * 1000
df.head(10) # table of top 10 players as an example
```

```python
plt.style.use('seaborn-notebook')
df.plot.scatter(x='play_count', y='rank', figsize=(10,5), grid=True, linewidth=2)
# It seems there is no correlation between the amount of plays a user has when compared to their rank
df[['play_count', 'rank']].corr()
```

```python
plt.style.use('fivethirtyeight')
df.plot.line(x='rank', y='performance', figsize=(10,4), grid=True, linewidth=4)
```

The line plot above compares the player's performanance points to the rank of the player. Since rank is calculated with performance points, there is a direct correlation. However, there is an apparent inverse relationship as seen by the gradually slowing slope. This means that the improvement from the top two players, for example, is much greater players #249 & 250.

```python
plt.style.use('seaborn-dark-palette')
dfscores = df[df['S'] <=8000][df['SS'] <=1000]
dfscores.drop(['username', 'play_count', 'performance'], axis=1, inplace=True)
dfscores.plot.scatter(x='A', y='accuracy', figsize=(10,5), grid=True, linewidth=2)
dfscores.plot.scatter(x='S', y='accuracy', figsize=(10,5), grid=True, linewidth=2)
dfscores.plot.scatter(x='SS', y='accuracy', figsize=(10,5), grid=True, linewidth=2)
dfscores[['accuracy', 'SS', 'S', 'A']].corr()
# After removing some outliers, it looks like the amount of 'SS' and 'S' may be correlated to the account's total accurac
```
