# Osu! Top 250 Players Dataset

### This dataset contains the top 250 ranked players in the rythm video game 'osu!', including their username, levels they played (play_count), total performance points (method of determining how good a player is), their average accuracy across levels (%), and the amount of the different accuracies they scored (SS [100%], S [95%], A [90%])

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
plt.style.use('ggplot')

df = pd.read_csv('./data/osu.csv', sep=',')
df['rank'] = df['rank'].str.replace('#','').astype(int)
df['accuracy'] = df['accuracy'].str.replace('%','').astype(float)
df['A'] = df['A'].apply(lambda x: (x*1000 if int(x) != x else x))
df['S'] = df['S'].apply(lambda x: (x*1000 if int(x) != x else x))
df['SS'] = df['SS'].apply(lambda x: (x*1000 if int(x) != x else x))
df['performance'] = df['performance'] * 1000
df['play_count'] = df['play_count'] * 1000
df.head(10) # table of top 10 players as an example
```

```python
df.plot.scatter(x='rank', y='play_count', figsize=(10,5), grid=True, color='Red', linewidth=2)
# It seems there is no correlation between the amount of plays a user has when compared to their rank
df[['play_count', 'rank']].corr()
```

This line plot compares the performanance points to the rank of the player. Since rank is calculated with performance points, there is a direct correlation. There is an apparent inverse relationship as seen by the gradually slowing slope.

```python
plt.style.use('fivethirtyeight')
df.plot.line(x='rank', y='performance', figsize=(10,5), grid=True, color='Red', linewidth=4)
```

```python
df.plot.scatter(x='accuracy', y='A', figsize=(10,5), grid=True, color='Red', linewidth=2)
df[df['S'] <=8000].plot.scatter(x='accuracy', y='S', figsize=(10,5), grid=True, color='Red', linewidth=2)
df[df['SS'] <=1000].plot.scatter(x='accuracy', y='SS', figsize=(10,5), grid=True, color='Red', linewidth=2)
df[['accuracy', 'SS', 'S', 'A']].corr()
# After removing some outliers, it looks like the amount of 'SS' and 'S' may be correlated to the account's total accurac
```
