2. Create and test a function that given a list of numbers will check whether each number is between 1 and 100, inclusive. If so, keep the number unchanged. If not, overwrite the value with 0. Return the modified list.

```python
def function2(nums):
    for i in range(len(nums)):
        if nums[i] < 1 or nums[i] > 100:
            nums[i] = 0
    return nums
function2([-1,0,1,2,98,99,100,101,102])
```

```python
class Book():
    def __init__(self, title, author, pages):
        print ("A book is created")
        self.title = title
        self.author = author
        self.pages = pages

    def __str__(self):
        return "Title:%s , author:%s, pages:%s " %(self.title, self.author, self.pages)

    def __len__(self):
        return self.pages

    def __del__(self):
        print ("A book is destroyed")
```
