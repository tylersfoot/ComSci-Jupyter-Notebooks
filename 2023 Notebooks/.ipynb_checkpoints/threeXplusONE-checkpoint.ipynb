```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import numpy as np
import pandas as pd
from matplotlib import cm
```

```python
import time
startTime = time.time()
Nrange = 200000
steps = 0
number = 0
maxsteps = 0
maxnumber = 0
listnums = []
liststeps = []
for i in range(1,Nrange):
    number = i
    while i != 1:
        if i % 2 == 0:
            i /= 2
        else:
            i = i*3 + 1
        steps += 1
    if steps >= maxsteps:
        maxsteps = steps
        maxnumber = number
    listnums+=[number]
    liststeps+=[steps]
    steps = 0
print(f'Max Number: {maxnumber} - {maxsteps} steps')

executionTime = (time.time() - startTime)
print('Execution time in seconds: ' + str(round(executionTime,2)))
# print(listnums)
# print(liststeps)
```

```python
df = pd.DataFrame({'number': listnums,
                   'steps': liststeps,})
df.set_index('number')

df.plot(x='number', y='steps', figsize=(15,10), grid=True)
```
